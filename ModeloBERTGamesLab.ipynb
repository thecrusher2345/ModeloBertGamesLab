```
from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments
from transformers import AutoTokenizer
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import pandas as pd
import torch
from transformers import EarlyStoppingCallback
import nlpaug.augmenter.word as naw
from googletrans import Translator
import random
```

```
print(f"Versión de PyTorch: {torch.__version__}")

# Cargar el archivo CSV completo
df = pd.read_csv('/content/MOCK_DATA.csv', encoding='utf_8_sig', delimiter=',')

# Replace NaN values
df['label'] = df['label'].fillna(0)

df['label'] = df['label'].astype(int)
```

```
# Filtrar filas con valores None en la columna 'text'
df = df.dropna(subset=['text'])
```

```
import nltk
nltk.download('averaged_perceptron_tagger_eng')
# Aplicar aumento de datos
aug = naw.SynonymAug(aug_src='wordnet')
df['augmented_text'] = df['text'].apply(lambda x: aug.augment(x))

translator = Translator()

def back_translate(text, src_lang='es', tgt_lang='en'):
    translated = translator.translate(text, src=src_lang, dest=tgt_lang).text
    back_translated = translator.translate(translated, src=tgt_lang, dest=src_lang).text
    return back_translated

def inject_noise(text):
    words = text.split()
    num_words = len(words)
    num_noisy = max(1, int(0.1 * num_words))  # Añadir ruido al 10% de las palabras
    noisy_indices = random.sample(range(num_words), num_noisy)
    for idx in noisy_indices: # Corrected the loop to iterate over noisy_indices directly
        words[idx] = ''.join(random.sample(words[idx], len(words[idx])))
    return ' '.join(words)

def random_deletion(text, p=0.1):
    words = text.split()
    if len(words) == 1:
        return text
    remaining_words = [word for word in words if random.uniform(0, 1) > p]
    if len(remaining_words) == 0:
        return random.choice(words)
    return ' '.join(remaining_words)

def random_case_change(text):
    return ' '.join([word.upper() if random.uniform(0, 1) > 0.5 else word.lower() for word in text.split()])

# Aplicar técnicas de aumento de datos y combinar los resultados
augmented_df = pd.concat([
    df[['text', 'label']],
    # Apply back_translate to each string in the 'text' column using applymap
    df[['text']].applymap(lambda x: back_translate(x)).rename(columns={'text': 'augmented_text'}).assign(label=df['label']),
    df[['text']].applymap(lambda x: inject_noise(x)).rename(columns={'text': 'augmented_text'}).assign(label=df['label']),
    df[['text']].applymap(lambda x: random_deletion(x)).rename(columns={'text': 'augmented_text'}).assign(label=df['label']),
    df[['text']].applymap(lambda x: random_case_change(x)).rename(columns={'text': 'augmented_text'}).assign(label=df['label'])
])
```

```
# prompt: haz que se imprima el numero de filas que resulto del dataset

print(f"Número de filas en el DataFrame aumentado: {len(augmented_df)}")
```

```
# Combinar el texto original y el aumentado
augmented_df = pd.concat([df[['text', 'label']], df[['augmented_text', 'label']].rename(columns={'augmented_text': 'text'})])
```

```
# Separar los datos en conjunto de entrenamiento (80%) y prueba (20%)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```
# Guardar los conjuntos de datos en archivos CSV separados
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```

```
# Cargar el tokenizer BETO
tokenizer = AutoTokenizer.from_pretrained('dccuchile/bert-base-spanish-wwm-cased')

# Cargar conjunto de datos separados
data_files = {"train": "train.csv", "test": "test.csv"}
dataset = load_dataset('csv', data_files=data_files)

# Preprocesar los textos
def preprocess_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True)
```

```
# Aplicar el preprocesamiento
encoded_dataset = dataset.map(preprocess_function, batched=True)

# Eliminar la columna de texto y renombrar la columna de etiquetas
encoded_dataset = encoded_dataset.remove_columns(['text'])
encoded_dataset = encoded_dataset.rename_column('label', 'labels')

# Filtrar filas con valores None en la columna 'labels'
encoded_dataset['train'] = encoded_dataset['train'].filter(lambda x: x['labels'] is not None)
encoded_dataset['test'] = encoded_dataset['test'].filter(lambda x: x['labels'] is not None)

# Convertir manualmente a tensores de PyTorch
encoded_dataset['train'] = encoded_dataset['train'].map(lambda x: {
    'input_ids': torch.tensor(x['input_ids']),
    'attention_mask': torch.tensor(x['attention_mask']),
    'labels': torch.tensor(x['labels'], dtype=torch.long)  # Convertir a tipo long
})

encoded_dataset['test'] = encoded_dataset['test'].map(lambda x: {
    'input_ids': torch.tensor(x['input_ids']),
    'attention_mask': torch.tensor(x['attention_mask']),
    'labels': torch.tensor(x['labels'], dtype=torch.long)  # Convertir a tipo long
})

print(encoded_dataset['train']['labels'][:10])
```

```
from transformers import EarlyStoppingCallback
# Cargar modelo preentrenado BETO para clasificación binaria
model = AutoModelForSequenceClassification.from_pretrained(
    'dccuchile/bert-base-spanish-wwm-cased',
    num_labels=2  # Número de clases (etiquetas)
)

# Definir función para calcular métricas
def compute_metrics(p):
    preds = p.predictions.argmax(-1)
    labels = p.label_ids
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    accuracy = accuracy_score(labels, preds)
    return {
        'accuracy': accuracy,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=10,  # Aumentar número de épocas
    per_device_train_batch_size=4,  # Ajustar tamaño del lote
    per_device_eval_batch_size=32,  # Ajustar tamaño del lote
    warmup_steps=500,
    weight_decay=0.02,  # Ajustar tasa de decaimiento del peso
    logging_dir='./logs',
    evaluation_strategy="epoch", # Evaluate at the end of each epoch
    save_strategy="epoch", # Save at the end of each epoch
    learning_rate=2e-5 ,  # Añadir tasa de aprendizaje
    load_best_model_at_end=True,  # To save and load the best performing model
    metric_for_best_model="f1", # Or another relevant metric like 'f1', 'accuracy'
    greater_is_better=True, # For eval_loss, lower is better hence False
    fp16=True,  # Use mixed precision
)

early_stopping_callback = EarlyStoppingCallback(early_stopping_patience=3) # You can adjust patience as needed

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test'],
    compute_metrics=compute_metrics,  # Añadir función de evaluación
    callbacks=[early_stopping_callback]
)
```

```
# Entrenar modelo
trainer.train()
```

```
# Evaluar modelo en conjunto de prueba
eval_results = trainer.evaluate()
print(eval_results)
```
